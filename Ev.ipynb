```python
df['Year'] = df['Year'].astype(int)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
categorical_columns = ['Month_Name', 'State', 'Vehicle_Class',
'Vehicle_Category', 'Vehicle_Type']
df[categorical_columns] = df[categorical_columns].astype('category')
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))  
plt.title('Yearly Analysis of EV Sales in India', fontsize=16)  
sns.lineplot(x='Year', y='EV_Sales_Quantity', data=df,
             marker='o', color='green')  
plt.xlabel('Year', fontsize=12)  
plt.ylabel('EV Sales', fontsize=12)
plt.grid(True)  
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))  
plt.title('Monthly Analysis of EV Sales in India', fontsize=18, fontweight='bold')
sns.lineplot(
    x='Month_Name',
    y='EV_Sales_Quantity',
    data=df,
    marker='D',        
    color='#4B0082',   
    linewidth=2.5      
)

plt.xlabel('Month', fontsize=14)
plt.ylabel('EV Sales', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)  
plt.grid(alpha=0.3)  
plt.tight_layout()   
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
plt.title('State-Wise Analysis of EV Sales', fontsize=18, fontweight='bold')

sns.barplot(
    y='State',
    x='EV_Sales_Quantity',
    data=df,
    hue='State',         
    palette='viridis',   
    dodge=False,
    legend=False         
)

plt.xlabel('EV Sales', fontsize=14)
plt.ylabel('States', fontsize=14)
plt.grid(axis='x', alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(24, 10)) 
sns.barplot(
    x='Vehicle_Class',
    y='EV_Sales_Quantity',
    data=df,
    hue='Vehicle_Class',
    palette='viridis',   
    dodge=False,
    legend=False
)

plt.title('Analysis by Vehicle Class', fontsize=24, fontweight='bold')
plt.xlabel('Vehicle Class', fontsize=18)
plt.ylabel('EV Sales', fontsize=18)
plt.xticks(rotation=90, ha='center', fontsize=16)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group data
category_sales = df.groupby('Vehicle_Category', observed=False)['EV_Sales_Quantity'].sum()
colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9', '#92A8D1', '#955251']

category_sales = category_sales.sort_values(ascending=False)

plt.figure(figsize=(6, 4))
bars = plt.bar(
    category_sales.index,
    category_sales.values,
    color=colors[:len(category_sales)]
)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{int(bar.get_height())}',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.title('EV Sales by Vehicle Category', fontsize=16, fontweight='bold')
plt.xlabel('Vehicle Category', fontsize=12)
plt.ylabel('EV Sales Quantity', fontsize=12)
plt.xticks(rotation=30)  
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

category_sales = df.groupby('Vehicle_Type', observed=False)['EV_Sales_Quantity'].sum()

plt.figure(figsize=(8, 5))
bars = plt.bar(
    category_sales.index,
    category_sales.values,
    color=['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9', '#92A8D1', '#955251']
)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height(),
             f'{int(bar.get_height())}',
             ha='center', va='bottom', fontsize=10)

plt.title('EV Sales by Vehicle Type (Clustered)', fontsize=14, fontweight='bold')
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('EV Sales Quantity', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
